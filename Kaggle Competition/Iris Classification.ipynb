En el DataFrame no hay valores perdidos.

Para detectar si hay outliers vamos a utilizar el método del Rango Intercuartílico (IQR).

```python
Q1_SepalLengthCm = iris['SepalLengthCm'].quantile(0.25)
print(Q1_SepalLengthCm)
```

```python
Q3_SepalLengthCm = iris['SepalLengthCm'].quantile(0.75)
print(Q3_SepalLengthCm)
```

```python
IQR_SepalLengthCm= Q3_SepalLengthCm - Q1_SepalLengthCm
print(IQR_SepalLengthCm)
```

```python
Lower_SepalLengthCm = Q1_SepalLengthCm - 1.5 * IQR_SepalLengthCm
print(Lower_SepalLengthCm)
```

```python
Upper_SepalLengthCm = Q3_SepalLengthCm + 1.5 * IQR_SepalLengthCm
print(Upper_SepalLengthCm)
```

```python
no_outliers_iris = iris[(iris['SepalLengthCm']>Lower_SepalLengthCm) & (iris['SepalLengthCm']<Upper_SepalLengthCm)]
no_outliers_iris
```

```python
Q1_SepalWidthCm = iris['SepalWidthCm'].quantile(0.25)
print(Q1_SepalWidthCm)
```

```python
Q3_SepalWidthCm = iris['SepalWidthCm'].quantile(0.75)
print(Q3_SepalWidthCm)
```

```python
IQR_SepalWidthCm= Q3_SepalWidthCm - Q1_SepalWidthCm
print(IQR_SepalWidthCm)
```

```python
Lower_SepalWidthCm = Q1_SepalWidthCm - 1.5 * IQR_SepalWidthCm
print(Lower_SepalWidthCm)
```

```python
Upper_SepalWidthCm = Q3_SepalWidthCm + 1.5 * IQR_SepalWidthCm
print(Upper_SepalWidthCm)
```

```python
no_outliers_iris = no_outliers_iris[(no_outliers_iris['SepalWidthCm']>Lower_SepalWidthCm) & (no_outliers_iris['SepalWidthCm']<Upper_SepalWidthCm)]
no_outliers_iris
```

```python
Q1_PetalLengthCm = iris['PetalLengthCm'].quantile(0.25)
print(Q1_SepalWidthCm)
```

```python
Q3_PetalLengthCm = iris['PetalLengthCm'].quantile(0.75)
print(Q3_PetalLengthCm)
```

```python
IQR_PetalLengthCm= Q3_PetalLengthCm - Q1_PetalLengthCm
print(IQR_PetalLengthCm)
```

```python
Lower_PetalLengthCm = Q1_PetalLengthCm - 1.5 * IQR_PetalLengthCm
print(Lower_PetalLengthCm)
```

```python
Upper_PetalLengthCm = Q3_PetalLengthCm + 1.5 * IQR_PetalLengthCm
print(Upper_PetalLengthCm)
```

```python
Q1_PetalWidthCm = iris['PetalWidthCm'].quantile(0.25)
print(Q1_PetalWidthCm)
```

```python
Q3_PetalWidthCm= iris['PetalWidthCm'].quantile(0.75)
print(Q3_PetalWidthCm)
```

```python
IQR_PetalWidthCm= Q3_PetalWidthCm - Q1_PetalWidthCm
print(IQR_PetalWidthCm)
```

**Conclusiones EDA**
- Iris-versicolor e Iris-virginica tienen un ancho de sépalo bastante similar. Las demás características parecen variar de una especie a otra.
- Existe una alta correlación positiva entre la longitud del sépalo y la anchura y longitud del pétalo.
- Existe una baja correlación negativa entre la longitud del sépalo y su anchura.
- También hay una cierta correlación negativa entre la anchura del sépalo y la anchura y longitud del pétalo.
- Además, existe una alta correlación entre la longitud y la anchura del pétalo.
- Finalmente, la especie Iris-versicolor y la especie Iris-virginica son bastante parecidas.

# Modeling

```python
X=iris_data.drop('Species', axis=1)
X.shape
```

```python
y=iris_data['Species']
y.shape
```

**Naive Bayes Classification**

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)

gnb=GaussianNB()
gnb.fit(X_train, y_train)
y_pred_gnb= gnb.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_gnb)
```

**Support Vector Machines**

```python
from sklearn.svm import SVC # 'Support Vector Classifier'

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)

svc=SVC()
svc.fit(X_train, y_train)
y_pred_svc=svc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_svc)
```

**Decision Tree Classifier**

```python
from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=3)

tree = DecisionTreeClassifier()
tree.fit(X_train, y_train)
y_pred_tree=tree.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_tree)
```

**Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
y_pred_rfc=rfc.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred_rfc)
```
